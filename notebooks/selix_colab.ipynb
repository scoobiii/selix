```python
class SELIX:
    TETO_1_DIGITO = 9.99
    JURO_REAL_MAXIMO = 5.0
    FOLGA_ROE = 0.95
    def __init__(self, inflacao=4.48, roe=31.23, selic_bacen=14.50):
        self.inflacao = inflacao
        self.roe = roe
        self.selic_bacen = selic_bacen
    def calcular_selix(self):
        teto_juro_real = self.inflacao + self.JURO_REAL_MAXIMO
        teto_roe = self.roe * self.FOLGA_ROE
        return round(min(self.TETO_1_DIGITO, teto_juro_real, teto_roe), 2)
    def diagnosticar(self):
        selix = self.calcular_selix()
        return {'selix_ideal': selix, 'juro_real_selix': round(selix - self.inflacao, 2)}

selix = SELIX()
res = selix.diagnosticar()
print(f'SELIX Ideal: {res["selix_ideal"]}%')
print(f'Juro real SELIX: {res["juro_real_selix"]}%')
```
