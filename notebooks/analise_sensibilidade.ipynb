# Análise de Sensibilidade da SELIX\nComo a SELIX varia com diferentes valores de inflação e ROE?

```
import numpy as np
import matplotlib.pyplot as plt

inflacao_range = np.linspace(3, 6, 10)
roe_range = np.linspace(20, 40, 10)

def selix(inflacao, roe):
    return min(9.99, inflacao+5.0, roe*0.95)

X, Y = np.meshgrid(inflacao_range, roe_range)
Z = np.vectorize(selix)(X, Y)

plt.contourf(X, Y, Z, levels=10, cmap='viridis')
plt.colorbar(label='SELIX (%)')
plt.xlabel('Inflação (%)')
plt.ylabel('ROE (%)')
plt.title('Sensibilidade da SELIX')
plt.show()
```
